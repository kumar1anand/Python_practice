# Loops in Python

**Why Loop**
- Execute same code repeatedly.

**While Loop**
- Executes while *condition* true.

**For Loop**
- Executes for each *item* in *iterable*.

````python
# Python syntax for while loop
# while condition:
#     code
````

````python
i=1
while i<10:
    print("the number" + str(i))
    i+=1   # or i=i+1
````

````python
number = int(input("Enter the number: "))
i = 1
while i < 11:
    print(number, "*", i, "=", number * i)
    i += 1
````

````python
x=2
while x<10:
    print(x)
    x=x+2
````

````python
#range(start,end,step)

print(list(range(1,10)))
````

````python
list1=[2,5,8,9,10,"bank"]

for item in list1:
    print(item)
````

**For Loop** : **Known** iterations.  
  ```python
  for item in iterable:
      # Code block
  ```
*iterable*: list, tuple, string, etc.

**While Loop**: **Unknown** iterations.  
  ```python
  while condition:
      # Code block
  ```
*condition*: boolean expression.

**Key Differences** is **For Loop** iterates a fixed number of times; **While Loop** iterates until condition is False.

**For Loop Usage** - Known iterations/sequence traversal; **While Loop Usage** - Uncertain iterations/condition-based.

````python
for deliveryCity in ["Kolkata", "Delhi", "Mumbai"]:
    print(deliveryCity)
````

````python
# range(start, stop, step) the range() function is used to generate a sequence of numbers. 

for productID in range(1,5):
    print(productID)

# Nested loops
for i in range(1, 5):
    for j in range(1, 5):
        print(i, j)
````

# Loop Control Statements

**Break**: Ends loop immediately.

**Continue**: Skips to next iteration.

**Pass**: Placeholder; does nothing.

````python
for i in range(1, 11):
    if i == 5:
        print("test before break stmt")
        break
        print("test after continue stmt")
    print(i)
````

````python
for i in range(1, 11):
    if i == 5:
        print("test before continue stmt")
        continue
        print("test after continue stmt")
    print(i)
````
