```python
# Clone - In cloning we create a copy of the list at a different memory address
# Shallow copy Vs Deep copy
# Shallow Copy: Copies the collection structure, but nested objects are shared (i.e., references to the same objects in memory).
# Deep Copy: Recursively copies all objects, ensuring that no references are shared between the original and the copy.
```

```python
# Method 1: Using slicing
original_list = [1, 2, [3, 4], 5]
shallow_copy = original_list[:]

print(id(original_list))
print(id(shallow_copy))

# Method 2: Using the list() constructor
shallow_copy = list(original_list)

# Method 3: Using the copy() method (for lists introduced in Python 3.3+)
shallow_copy = original_list.copy()

print(shallow_copy)
```

```python
shallow_copy[2][0] = 100

print(shallow_copy)
print(original_list)

shallow_copy[1] = 200

print(shallow_copy)
print(original_list)
```

```python
#Example illustrating shallow vs. deep copy

import copy

original_list = [1, 2, [3, 4]]
shallow_copy = original_list.copy() # Shallow
deep_copy = copy.deepcopy(original_list) # Deep

# Modifying a nested object in the shallow copy
shallow_copy[2][0] = 100

# Modifying a nested object in the deep copy
deep_copy[2][1] = 200

print("Original:", original_list)  # Affects shallow copy but not deep copy
print("Shallow Copy:", shallow_copy)
print("Deep Copy:", deep_copy)
```

```python
lst1=[]
dict1 = {}   # dictionary is mutable so the same object get updated in loop

for i in range(5):
    key = "key" + str(i)
    dict1[key] = i
    lst1.append(dict1)

print(lst1)
```

```python
lst2=[]

for i in range(5):
    dict2 = {}      # Create every time as it is mutable
    key = "key" + str(i)
    dict2[key] = i
    lst2.append(dict2)

print(lst2)
```
