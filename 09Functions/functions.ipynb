```python
### Variable Length Arguments
## Positional And Keywords arguments

def print_numbers(*arg):
    for number in arg:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"anytext")
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Prudvi",age="32",country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Prudvi",name="Prudvi",age="32",country="India")
```

```python
### Return statements
def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
### Return multiple parameters
def multiply(a,b):
    return a*b,a

as_single_return= multiply(2,3)
mul,a= multiply(2,3)

print(as_single_return)
print(mul)
print(a)
```

```python
def square(num):
    return num*num

def mul(num):
    return num*10

mul(9)


def highorderfun(func,num):
    print(func(num))


highorderfun(square,4) 
highorderfun(mul,78)   
```

```python
def addsectohttp(func):
    def wrapper(param2):
        changeddata= param2+" Security"
        func(changeddata)
    return wrapper
```

```python
@addsectohttp
def callhttp(param1):
    print(param1)


# callling the function
callhttp("data") 
```
