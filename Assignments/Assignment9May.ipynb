Write a Python program using reduce() to find the product of all numbers in a list.
Example:
Input: [1,2,3,4]
Output: 24

```python
from functools import reduce

nums = [1, 2, 3, 4]
result = reduce(lambda x,y: x*y, nums)
print(result)
```
