**No Var Decl in Python**

- Assign val directly; Python auto-assign the data type at runtime.

```python
# Static Typing (Java) at compile time
int a;
a=10;
```

```python
# Dynamic Binding Example
a = 5          # Assign integer
print(a)

a = 'nitish'   # Reassign string
print(a)
```

```python
# Special Syntax
a = 5; b = 6; c = 7

print(a,b,c)
```

# Literals

Literals = basic **data values** for variables.

Types in Python:
- *Numeric literals*
- *String literals*
- *Boolean literals*
- *Special literals*

```python
# Numeric Literal

# Integer Literals
b = 100     # Decimal

a = 0b1010  # Binary
d = 0x12c   # Hex

# Float Literals
f1 = 10.5
f2 = 1.5e2   # e2 = 10^2
f3 = 1.5e-3

# Complex Literal
x = 3.14j

print(a, b, c, d)
print(f1, f2, f3)
print(x, x.imag, x.real)
```

```python
# Special Literal
a= None
print(a)
```

```python
# string literal

strval= 'This is a string'

# boolean literal

bval= True
print(bval)
```
