```python
def login_required(func):
    def wrapper(*args, **kwargs):
        user_logged_in = True
        if user_logged_in:
            print("login successful")
        else:
            print("login failed")
            return "Access Denied"
        return func(*args, **kwargs)
    return wrapper
```

```python
def secure_function(func):
    def wrapper(*args, **kwargs):
        for arg in args:
            print(f"Checking argument: {arg}")
        # Here you can add your security checks
        print("Performing security checks...")
        # If checks pass, call the original function
        for key,value in kwargs.items():
            print(f"Checking keyword argument: {key}={value}")
        return func(*args, **kwargs)
    return wrapper
```

```python
@login_required
@secure_function
def getDataFromDB(args, **kwargs):
    print(f"Executing query: {args} with parameters {kwargs}")
    # Simulate database access
    return "Data from database"
```
