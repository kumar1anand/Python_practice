# Keywords

**Python is case sensitive**

- *Keywords*: reserved words, special meanings
- Cannot use as **variable names**

```python
# Python has 35 keywords.
import keyword
print(keyword.kwlist)
 
```

# Identifiers

**Identifier** are name for variables, functions, class, module, etc.

**Rules**:
- Start with letter or `_`
- Followed by letters, digits, `_`
- Not a Python keyword

```python
name = 'Saurabh'
print(name)
```

```python
_ = 'Saurabh'
print(_)
```

```python
first_name = 'Saurabh'
print(first_name)
```

```python
False = 'Saurabh' # False is keyword which is reserved for python program
print(False)
```
